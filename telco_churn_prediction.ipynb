```python
df = df.replace(to_replace={'International plan': {'No':0, 'Yes':1}})
df = df.replace(to_replace={'Voice mail plan': {'No':0, 'Yes':1}})
```

```python
df = df.rename(columns={'Total day minutes': 'total_day_minutes'}) 
#df = df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'})
```

```python
df = df.astype({"Number vmail messages":'int64'})
```

```python
import statsmodels.formula.api as smf

model = smf.logit(formula='Churn ~ total_day_minutes', data = df)

results = model.fit()
results.summary()
```

```python
import math
X = pd.Series(range(850))
intercept = results.params['Intercept']
slope = results.params['total_day_minutes']
p_y = 1 / (1 + math.e**-(intercept + (slope * X)))

X_df = pd.DataFrame(X)
X_df.columns = ['total_day_minutes']

p_y = results.predict(X_df)
p_y
```
