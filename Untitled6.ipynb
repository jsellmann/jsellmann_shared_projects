```python
def clean_features(features_train, features_test): 
    ''''This Function cleans tran- and test features that were created after a train_test split.'''

    #1st. Change daytime categories of features: 
    features_train.loc[:, 'PurchDate'] = pd.to_datetime(features_train.loc[:, 'PurchDate'])#, unit='s')
    features_train['Date'] = pd.to_datetime(features_train['PurchDate'])
    features_train['Year'] = features_train.Date.dt.year
    features_train['Month'] = features_train.Date.dt.month
    features_train['Day'] = features_train.Date.dt.day
    features_train = features_train.drop('PurchDate', axis = 1)
    features_train = features_train.drop('Date', axis = 1)
    
    
    features_test.loc[:, 'PurchDate'] = pd.to_datetime(features_test.loc[:, 'PurchDate'], unit='s')
    features_test['Date'] = pd.to_datetime(features_test['PurchDate'])
    features_test['Year'] = features_test.Date.dt.year
    features_test['Month'] = features_test.Date.dt.month
    features_test['Day'] = features_test.Date.dt.day
    features_test = features_test.drop('PurchDate', axis = 1)
    features_test = features_test.drop('Date', axis = 1) 
    
    return features_train, features_test
```

```python
#This is how it works with a column transformer! 

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline 
from sklearn.compose import ColumnTransformer 
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
col_transformer = ColumnTransformer(transformers = [
                                            ('ohe', OneHotEncoder(handle_unknown = 'ignore'), cat_cols), 
                                            ('scaler', StandardScaler(), num_cols)
])
```

```python
log_reg_pipeline = Pipeline(steps = [
                            ('col_transformer', col_transformer), 
                            ('model', LogisticRegression(random_state = 42, 
                                             C = 1e42, 
                                             max_iter = 1e4, 
                                             class_weight = 'balanced'))
])
```

```python
def pca_function(features_test): ######IT IS TRANSFORMED!!!!
    '''This function performs a Principal Component Analysis on highly correlated features of the Mondaycar Dataset'''
    
    from sklearn.pipeline import Pipeline
    from sklearn.preprocessing import StandardScaler
    from sklearn.decomposition import PCA

    col_correlated = ['MMRAcquisitionAuctionAveragePrice',    
                  'MMRAcquisitionAuctionCleanPrice',      
                  'MMRAcquisitionRetailAveragePrice',     
                  'MMRAcquisitonRetailCleanPrice', 
                  'MMRCurrentAuctionAveragePrice',        
                  'MMRCurrentAuctionCleanPrice',          
                  'MMRCurrentRetailAveragePrice',         
                  'MMRCurrentRetailCleanPrice']  

    standardizer = StandardScaler()
    pca=PCA(n_components = 0.8) #I chose 80%, because I want to have a nice overview of the features without
                            #loosing too many. 
    std_pca = Pipeline([('scale', standardizer), ('pca', pca)])

    price_std_pca = std_pca.fit_transform(features_test[col_correlated])

    features_test = features_test.drop(col_correlated, axis = 1)

    features_test['pca_all_price_0'] = price_std_pca[:, 0]
    features_test['pca_all_price_1'] = price_std_pca[:, -1]
    
    
    return features_test
```
